```python
# Definition of functions for each unit operation

# Sterilization mass balance
def sterilization_mass_balance(feedstock_inflow, water_inflow, steam_inflow):
    total_inflow = feedstock_inflow + water_inflow + steam_inflow
    return total_inflow

# Sterilization energy balance
def sterilization_energy_balance(steam_energy):
    return steam_energy

# Overall mass balance
def overall_mass_balance(total_inflow, total_outflow):
    return total_inflow - total_outflow

# Overall energy flows
def overall_energy_flows(feedstock_energy, steam_energy, distillation_feed_preheat_energy, reboiler_duty, cotreatment_power, ethanol_energy, residue_energy, post_sterilization_cooling_energy, condenser_duty):
    total_energy_in = feedstock_energy + steam_energy + distillation_feed_preheat_energy + reboiler_duty + cotreatment_power
    total_energy_out = ethanol_energy + residue_energy + post_sterilization_cooling_energy + condenser_duty
    return total_energy_in, total_energy_out

# Overall process efficiency
def process_efficiency(ethanol_energy, residue_energy, total_energy_in):
    return (ethanol_energy + residue_energy) / total_energy_in

# Define parameters
feedstock_inflow = 570  # kg/hr
water_inflow = 1429  # kg/hr
steam_inflow = 258  # kg/hr

# Perform calculations
total_inflow_sterilization = sterilization_mass_balance(feedstock_inflow, water_inflow, steam_inflow)
steam_energy_sterilization = sterilization_energy_balance(steam_inflow)

# Overall mass balance
total_outflow = 98 + 108 + 2050  # kg/hr
overall_mass_balance_result = overall_mass_balance(total_inflow_sterilization, total_outflow)


# Overall process efficiency
efficiency_result = process_efficiency(ethanol_energy, residue_energy, overall_energy_in)

# Results
print("Total Inflow Sterilization:", total_inflow_sterilization, "kg/hr")
print("Steam Energy Sterilization:", steam_energy_sterilization, "kJ/hr")
print("Total Outflow:", total_outflow, "kg/hr")
print("Overall Mass Balance:", overall_mass_balance_result, "kg/hr")
print("Overall Energy In:", overall_energy_in, "kJ/hr")
print("Overall Energy Out:", overall_energy_out, "kJ/hr")
print("Efficiency Result:", efficiency_result)



# Total mass flow in and out
total_mass_in = sum(mass_flow_values[:3])  # Feedstock + Water + Steam
total_mass_out = sum(mass_flow_values[3:])  # Total Out

# Total energy flow in and out
total_energy_in = sum(energy_flow_values[:5])  # Feedstock + Steam energy + Distillation feed preheat + Reboiler duty + Co-treatment power
total_energy_out = sum(energy_flow_values[5:])  # Hydrous ethanol + Wet residue + Post-sterilization cooling + Condenser duty

# Overall process efficiency
overall_efficiency = (energy_flow_values[5] + energy_flow_values[6]) / total_energy_in

# Results
print("Total Mass In:", total_mass_in, "kg/hr")
print("Total Mass Out:", total_mass_out, "kg/hr")
print("Total Energy In:", total_energy_in, "kJ/hr")
print("Total Energy Out:", total_energy_out, "kJ/hr")
print("Overall Process Efficiency:", overall_efficiency)
```
