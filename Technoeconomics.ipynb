```python
# Given financial parameters
discount_rate = 0.15

# Extracted information from Table C
TCI = 2169992
WC = 325499
variable_cost_per_year = 265095
fixed_cost_per_year = 143484
annual_hydrous_ethanol_gallons = 285307

# Define functions for calculating cash flows
def calculate_annual_depreciation(year):
    if year <= 10:  # Depreciation over 10 years
        return TCI / 10
    else:
        return 0

def calculate_annual_tax(year):
    earnings_before_tax = calculate_annual_production(year) - calculate_annual_variable_cost(year) - \
                          calculate_annual_fixed_cost(year) - calculate_annual_depreciation(year)
    return earnings_before_tax * 0.21

def calculate_annual_cash_flow(year):
    production_revenue = calculate_annual_production(year)
    variable_cost = calculate_annual_variable_cost(year)
    fixed_cost = calculate_annual_fixed_cost(year)
    depreciation = calculate_annual_depreciation(year)
    tax = calculate_annual_tax(year)
    cash_flow = production_revenue - variable_cost - fixed_cost - depreciation - tax
    return cash_flow

# Calculate cash flows for each year
cash_flows = [calculate_annual_cash_flow(year) for year in range(1, 13)]

# Calculate cumulative present value
discount_factors = [(1 / (1 + discount_rate) ** n) for n in range(1, 13)]
cumulative_pv = [sum(cash_flows[:i]) * df for i, df in enumerate(discount_factors, start=1)]

# Display the results
for i, cf in enumerate(cash_flows, start=1):
    print(f"{i} {cf:.2f} {cumulative_pv[i-1]:.2f}")
```

```python
import numpy as np
from numpy_financial import npv
from scipy.optimize import fsolve

# Given financial parameters
hydrous_ethanol_price_per_gallon = 4.00
wet_residue_price_per_kg = 0.01
project_life = 12
construction_period = 1
production_capacity_year2 = 0.50
production_capacity_years3_12 = 1.00
depreciation_years = 10
income_tax_rate = 0.21
discount_rate = 0.15

# Extracted information from Table C
annual_operating_hours = 8000
TCI = 2495491
TDC = 1972720
WC = 325499
FCI = TCI - WC
variable_cost_per_year = 265095
fixed_cost_per_year = 143484
annual_hydrous_ethanol_gallons = 285307
annual_wet_residue_kg = 16403772

# Define functions for calculating cash flows
def calculate_annual_production(year):
    if year == 1:
        return 0  # Construction period
    elif year == 2:
        return annual_hydrous_ethanol_gallons * production_capacity_year2 * hydrous_ethanol_price_per_gallon
    elif 3 <= year <= project_life:
        return annual_hydrous_ethanol_gallons * production_capacity_years3_12 * hydrous_ethanol_price_per_gallon

def calculate_annual_variable_cost(year):
    if year == 1:
        return 0  # Construction period
    else:
        return variable_cost_per_year

def calculate_annual_fixed_cost(year):
    if year == 1:
        return 0  # Construction period
    else:
        return fixed_cost_per_year

def calculate_annual_depreciation(year):
    if year <= depreciation_years:
        return TDC / depreciation_years
    else:
        return 0

def calculate_annual_tax(year):
    earnings_before_tax = calculate_annual_production(year) - calculate_annual_variable_cost(year) - \
                          calculate_annual_fixed_cost(year) - calculate_annual_depreciation(year)
    return max(0, earnings_before_tax * income_tax_rate)

def calculate_annual_cash_flow(year):
    production_revenue = calculate_annual_production(year)
    variable_cost = calculate_annual_variable_cost(year)
    fixed_cost = calculate_annual_fixed_cost(year)
    depreciation = calculate_annual_depreciation(year)
    tax = calculate_annual_tax(year)
    cash_flow = production_revenue - variable_cost - fixed_cost - depreciation - tax
    return cash_flow

# Calculate cash flows for each year
cash_flows = [calculate_annual_cash_flow(year) for year in range(1, project_life + 1)]

# Calculate NPV
npv_value = npv(discount_rate, cash_flows)

# Calculate IRR
irr_value = fsolve(lambda r: npv_value + TCI * r - cash_flows[0], 0.1)[0] * 100

# Calculate minimum selling price for 15% IRR
minimum_selling_price = fsolve(lambda p: npv_value + (cash_flows[2] - TCI * discount_rate) / ((1 + p) ** 2), 4)[0]

# Calculate payback period
payback_period = next(i for i, cf in enumerate(cash_flows, start=1) if cf >= 0)
cash_flows
```
